```python
team_ids = pd.read_csv('ncaa-team-id-edited.csv')
team_ids.head(10)
```

```python
past_results = pd.read_csv('TourneyCompactResults0216.csv')
past_results.head()
```

```python
winteam = []
losteam = []

for i, row in past_results.iterrows():
    for j, row2 in team_ids.iterrows():
        if row[1] == row2[1]:
            winteam.append(row2[2])
        if row[2] == row2[1]:
            losteam.append(row2[2]) 
        else:
            continue 
```

```python
year = past_results['Season'].values.tolist()   

temp_data = pd.DataFrame({'Year':year,'WinTeam':winteam,'LoseTeam':losteam})

temp_data = temp_data[['Year','WinTeam','LoseTeam']]
```

```python
year_array = [2002,2003,2004,2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016]
win = []
for i in range(0,len(year_array)):
    
    temp_teams = data.loc[data['Year'] == year_array[i]]
    temp_wl = temp_data.loc[temp_data['Year'] == year_array[i]]
    
    for j , row in temp_wl.iterrows():
        win_team = row[1]
        lose_team = row[2]
        #print win_team , lose_team
        wlogic = temp_teams['Team'] == win_team
        llogic = temp_teams['Team'] == lose_team
        wrank = temp_teams[wlogic].iloc[0].name
        lrank = temp_teams[llogic].iloc[0].name
        wdata = pd.DataFrame.as_matrix(temp_teams[wlogic])
        ldata = pd.DataFrame.as_matrix(temp_teams[llogic])

        if wdata[0,1] < ldata[0,1]:
            win.append(1)
        elif wdata[0,1] > ldata[0,1]:
            win.append(0)
        elif wdata[0,1] == ldata[0,1] and wrank < lrank:
            win.append(1)
        else:
            win.append(0)
        
```

```python
tourn_win_data = pd.DataFrame({'Year':year,'WinTeam':winteam,'LoseTeam':losteam,'Win':win})
tourn_win_data = tourn_win_data[['Year','WinTeam','LoseTeam','Win']]
tourn_win_data.to_csv('tourney-results-0216.csv')
```
